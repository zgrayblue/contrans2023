```python
import numpy as np
import pandas as pd
import requests
import json
import os
opensecretsapi = os.getenv('opensecretsapi')
r = requests.get('https://httpbin.org/user-agent')
useragent = json.loads(r.text)['user-agent']
```

## Problem 2

```python
states = ['AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME', 'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
```

```python
root = 'http://www.opensecrets.org/api/'
headers = {'User-Agent': useragent,
          'From': 'pdy2bw@virginia.edu'}
all_state_df = pd.DataFrame()
for state in states:
    parameters = {'method' : 'getLegislators',
             'apikey' : opensecretsapi,
             'id' : state,
             'output' : 'json'}
    r = requests.get(root, params = parameters, headers = headers)
    myjson = json.loads(r.text)['response']['legislator']

    legislators = []
    for legislator in myjson:
        attributes = legislator['@attributes']
        legislators.append(attributes)
    state_df = pd.DataFrame.from_dict(legislators)
    all_state_df = pd.concat([all_state_df, state_df])

all_state_df
```

```python
cid = all_state_df.query("lastname == 'Good'")['cid']
bh_cid = cid[4]
bh_cid
```

```python
root = 'http://www.opensecrets.org/api/'
headers = {'User-Agent': useragent,
          'From': 'pdy2bw@virginia.edu'}
parameters = {'method' : 'candContrib',
             'apikey' : opensecretsapi,
             'cid' : bh_cid,
             'output' : 'json'}
r = requests.get(root, params = parameters, headers = headers)
myjson = json.loads(r.text)['response']['contributors']['contributor']

contributors = []
for c in myjson:
    contributors.append(c['@attributes'])

contributors_df = pd.DataFrame.from_dict(contributors)
contributors_df
```
